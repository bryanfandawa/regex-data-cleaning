Now working on with rows. When we want to do a str operation and regex, we need to convert it to String. so astype(Str).str.replace(...) if the datatype is not yet a string. In this case its already a string, so its unecessary to convert it to a string.

```python
df.info()
```

```python
df["actual_gross"] = df["actual_gross"].str.replace(r'[^0-9.]', '', regex = True).astype(float)
```

```python
df["adjusted_gross_2022(dollars)"] = df["adjusted_gross_2022(dollars)"].str.replace(r'[^0-9.]', '', regex = True).astype(float)
```

```python
df["average_gross"] = df["average_gross"].str.replace(r'[^0-9.]', '', regex = True).astype(float)
```

We also need to remove accents and diactritics from characters. Such as Beyoncé. We need to first import unicodedata

```python
import unicodedata
def rem_accents(text):
  text = unicodedata.normalize('NFKD', text)
  text = re.sub(r'[\u0300-\u036f]', '', text)
  return text

df["artist"] = df["artist"].apply(rem_accents)
```

Let us remove all the unique symbols from tour_title column

```python
df["tour_title"].unique()
```

```python
def remove_symbols(text):
  text = re.sub(r'\[.*?\]', '', text)
  text = re.sub(r'[†‡*]', '', text)
  return text.strip()

df["tour_title"] = df["tour_title"].astype(str).apply(remove_symbols)
```

Cleaning for "years"

```python
df["years"].unique()
```

```python
df["years"] = df["years"].str.strip()
df["years"] = df["years"].str.replace('–', '-')
```

Let us try to make a new column named year_start and year_end

```python
df[["year_start", "year_end"]] = df["years"].str.extract(r'(\d{4})-?(\d{4})?')
#the regex code above for \d means all digits from 0-9
#{4} means exactly 4 numbers, closed by the parenthesis (group 1)
#-? means there could be a - or non, giving a heads up
#same for the second group that ends with a ? because there could be a year.
df["year_end"] = df["year_end"].fillna(df["year_start"])
```
